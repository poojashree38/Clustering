# **K-Mean**

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
Airline=pd.read_excel('EastWestAirlines.xlsx')
Airline
```

```python
Airline.info()
```

```python
def minmax_func(i):
  x=(i-i.min())/(i.max()-i.min())
  return(x)
```

```python
x= minmax_func(Airline.iloc[:,:])
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler_airline=scaler.fit_transform(Airline.iloc[:,:])
scaler_airline
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit_predict(x)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
clusters_new=KMeans(4,random_state=0)
clusters_new.fit(x)
```

```python
clusters_new.labels_
```

```python
Airline['h_clusterid']=clusters_new.labels_
Airline
```

```python
Airline[Airline['h_clusterid']==0]
```

```python
Airline[Airline['h_clusterid']==1]
```

```python
Airline[Airline['h_clusterid']==2]
```

```python
Airline[Airline['h_clusterid']==3]
```

```python
Airline[Airline['h_clusterid']==4]
```

```python
clusters_new.cluster_centers_
```

# **DBSCAN**

```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from matplotlib import pyplot as plt
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
Airline=pd.read_excel('EastWestAirlines.xlsx')
Airline
```

```python
data=Airline.values
data
```

```python
from sklearn.preprocessing import StandardScaler
stscaler=StandardScaler()
x=stscaler.fit_transform(data)
x
```

```python
dbscan=DBSCAN(eps=3,min_samples=6)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
Airline['Clusterid']=dbscan.labels_
Airline
```

```python
Airline['Clusterid'].value_counts()
```

```python
(len(Airline[Airline['Clusterid']==-1])/len(Airline))*100
```

```python
from sklearn import metrics
X=metrics.silhouette_score(x,dbscan.labels_)
X
```

```python
Airline.groupby(Airline['Clusterid']).mean()
```
